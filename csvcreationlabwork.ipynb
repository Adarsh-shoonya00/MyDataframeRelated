```python
import numpy as np
import pandas as pd
import random
```

```python
li = []
for i in range(1000):
    li.append(random.randint(10, 100))
    
li = np.array(li).reshape(100,10)
df = pd.DataFrame(li)
df.to_csv('lab2.csv') 
myfile = pd.read_csv('lab2.csv')
import csv
with open('lab2.csv','r',newline='') as f:
  readfile = csv.reader(f,delimiter=',')
  for i in readfile:
    print(i)
print('max :',li.max(axis=0))
print('min:',li.min(axis =0))

variance = df.var()
covariance= df.cov()
print(variance ,covariance)
```
